```python
def extraction_text_from_pdf(file_path, file_names):
    words = []
    bboxes = []

    for i,file_name in enumerate(file_names):
        words.append([])
        bboxes.append([])

        doc = fitz.open(f"{file_path}{file_name}.pdf")
        for page in doc:
            data = page.get_text("words")
            print("page")

            width = page.mediabox[2]
            height = page.mediabox[3]

            for d in data:
                words[i].append(d[4].lower())
                bbox = normalize_bbox(list(d[0:4]), [width, height])
                bboxes[i].append(bbox)
        
    return words, bboxes

def normalize_bbox(bbox, size):
    x0 = int(1000*bbox[0] / size[0])
    y0 = int(1000*bbox[1] /size[1])
    x1 = int(1000*bbox[2] / size[0])
    y1 = int(1000*bbox[3] / size[1])
    if x0 > 1000: x0 = 1000
    if y0 > 1000: y0 = 1000
    if x1 > 1000: x1 = 1000
    if y1 > 1000: y1 = 1000
    if x0 < 0: x0 = 0
    if y0 < 0: y0 = 0
    if x1 < 0: x1 = 0
    if y1 < 0: y1 = 0
    return [x0, y0, x1, y1]
    # return [
    #     int(1000 * bbox[0] / size[0]),
    #     int(1000 * bbox[1] / size[1]),
    #     int(1000 * bbox[2] / size[0]),
    #     int(1000 * bbox[3] / size[1]),
    # ]
```

```python
file_path = "../../../datasets/pdfs/train/"
filenames = ["PMC2678976_00003", "PMC5954792_00004"]
words, bboxes = extraction_text_from_pdf(file_path, filenames)
```
